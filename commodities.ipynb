```python
df = pd.read_csv('heatmap_data.csv'  , low_memory=False   )
```

```python
df.head()
```

```python
df.dropna(subset=['latitude'], how='all', inplace=True)
df.isna().sum()
```

```python

df_wheat  =  df[df['Commodity']=='Wheat']
df_rye    = df[df['Commodity']=='Rye']
df_barley = df[df['Commodity']=='Barley']
df_wine   = df[df['Commodity']=='Wine']
df_oats   = df[df['Commodity']=='Oats']
```

```python
from folium.plugins import Draw

import pandas as pd
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
,
m = folium.Map(world_copy_jump=False, location= [43.8525872,-12.8611109] ,
    no_wrap=True , tiles= 'cartodbpositron', zoom_start=1.3 )


steps=10
colormap = branca.colormap.linear.YlOrRd_07.scale(0, 1).to_step(steps)
#adding a title to the color bar
#colormap.caption = 'The more red it, is the more complaints there are'
gradient_map=defaultdict(dict)
for i in range(steps):
    gradient_map[1/steps*i] = colormap.rgb_hex_str(1/steps*i)
colormap.add_to(m)


coordinates_1 =[]
for la,lo in zip( df['latitude'],df['longitude']):
    coordinates_1.append([la,lo])
    
coordinates_2 =[]
for la,lo in zip( df_wheat['latitude'],df_wheat['longitude']):
    coordinates_2.append([la,lo])
    
coordinates_3 =[]
for la,lo in zip( df_rye['latitude'],df_rye['longitude']):
    coordinates_3.append([la,lo])
    
coordinates_4 =[]
for la,lo in zip( df_barley['latitude'],df_barley['longitude']):
    coordinates_4.append([la,lo])
    
coordinates_5 =[]
for la,lo in zip( df_wine['latitude'],df_wine['longitude']):
    coordinates_5.append([la,lo])
    
coordinates_6 =[]
for la,lo in zip( df_oats['latitude'],df_oats['longitude']):
    coordinates_6.append([la,lo])


#creating a Marker for each point in df_sample.  Each point will get a popup with their zip

m.add_child(plugins.HeatMap(coordinates_1, gradient = gradient_map, max_zoom=17, radius=10, blur=10, overlay = True , name = "countries"))
m.add_child(plugins.HeatMap(coordinates_2, gradient = gradient_map, max_zoom=17, radius=10, blur=10, overlay = True ,show = False, name = "wheat"))
m.add_child(plugins.HeatMap(coordinates_3, gradient = gradient_map, max_zoom=17, radius=10, blur=10, overlay = True ,show = False, name = "rye"))
m.add_child(plugins.HeatMap(coordinates_4, gradient = gradient_map, max_zoom=17, radius=10, blur=10, overlay = True , show = False,name = "barley"))
m.add_child(plugins.HeatMap(coordinates_5, gradient = gradient_map, max_zoom=17, radius=10, blur=10, overlay = True ,show = False, name = "wine"))
m.add_child(plugins.HeatMap(coordinates_6, gradient = gradient_map, max_zoom=17, radius=10, blur=10, overlay = True ,show = False, name = "oats"))







#adding diffrent styles to the map
folium.TileLayer('openstreetmap').add_to(m)
folium.TileLayer('stamentoner').add_to(m)
folium.LayerControl(collapsed=True).add_to(m)

title_html = '''
             <h3 align="center" style="font-size:20px"><b> HeatMap of Top 5 Commodities </b></h3>
             '''
m.get_root().html.add_child(folium.Element(title_html))



m.save("HeatMap of Top 5 Commodities.html")

```
